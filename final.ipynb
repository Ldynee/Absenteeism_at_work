# *2.*Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier

# Tạo và huấn luyện mô hình
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)

# Dự đoán và tính toán độ chính xác
rf_y_pred = rf_classifier.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_y_pred)
print('Random Forest Accuracy:', rf_accuracy)

# Báo cáo kết quả
rf_classification_report = classification_report(y_test, rf_y_pred)
print('Random Forest Classification Report:\n', rf_classification_report)
```

* Mô hình Random Forest có độ chính xác (accuracy) rất cao là 0.99, cho thấy mô hình rất hiệu quả trong việc dự đoán.

```python
import pandas as pd

# Lấy danh sách các nhân viên duy nhất từ cột 'ID'
employees = data['ID'].unique()

# Sắp xếp id theo thứ tự từ bé đến lớn
employees = sorted(employees)

# Tạo một DataFrame mới để lưu trữ kết quả dự đoán
predictions = pd.DataFrame(columns=['ID', 'Absenteeism prediction'])

# Duyệt qua từng nhân viên và dự đoán thời gian nghỉ việc
for emp_id in employees:
    # Lấy dữ liệu của nhân viên emp_id
    X_emp = data[data['ID'] == emp_id][X.columns]  # X.columns là danh sách các đặc trưng đã sử dụng

    # Dự đoán sử dụng mô hình Random Forest đã huấn luyện (rf_classifier)
    prediction = rf_classifier.predict(X_emp)

    # Tạo DataFrame tạm thời để lưu kết quả dự đoán
    df_temp = pd.DataFrame({'ID': [emp_id], 'Absenteeism prediction': [prediction[0]]})

    # Thêm df_temp vào predictions bằng phương thức concat
    predictions = pd.concat([predictions, df_temp], ignore_index=True)

# Hiển thị kết quả
print(predictions)

```

```python
from tabulate import tabulate
table = predictions.values.tolist()

# Map predictions to 'Absenteeism' or 'No Absenteeism'
predictions['Absenteeism prediction'] = predictions['Absenteeism prediction'].map({1: 'Nghỉ việc', 0: 'Không nghỉ việc'})
# Header for the table
header = ['ID employee', 'Absenteeism prediction']

# Print the formatted table using tabulate
print(tabulate(table, headers=header, tablefmt='fancy_grid'))
```
